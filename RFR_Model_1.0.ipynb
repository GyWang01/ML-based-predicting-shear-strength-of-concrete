```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 导入数据
dataset_ = pd.read_csv('data_1.0.csv',index_col=0)
```

```python
# 输出数据预览
dataset_.head()
```

```python
dataset_.info()
```

```python
#特征分布预览
dataset_.describe([0.01,0.1,0.25,0.5,0.75,0.9,0.99]).T
```

```python
# 准备训练数据
# 特征：第一列至倒数第二列
# 标签：最后一列
X = dataset_.iloc[:, 0:-1].values
y = dataset_.iloc[:, -1].values
```

```python
#标准化特征矩阵
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #实例化
scaler.fit(X) #fit，生成均值和方差
X_std = scaler.transform(X) #通过接口导出结果
```

```python
# 将数据分为训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std,
                                                    y,
                                                    test_size=0.1,
                                                    random_state=0)
```

```python
#绘制n_estimator学习曲线
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
superpa = []
for i in range(200):
    rfr = RandomForestRegressor(n_estimators=i+1,n_jobs=-1,random_state=4)
    rfr_s = cross_val_score(rfr,X_std,y,cv=10).mean()
    superpa.append(rfr_s)
print(max(superpa),superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1,201),superpa)
plt.show()
```

```python
# 训练随机森林解决回归问题
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion='friedman_mse',n_estimators = 93,random_state=4)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
score = regressor.score(X_test, y_test)
score
```

```python
#交叉验证分数
from sklearn.model_selection import cross_val_score
rfr_s = cross_val_score(regressor,X_std,y,cv=10).mean()
rfr_s
```

```python
# 评估回归性能
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:',
      np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R^2:', metrics.r2_score(y_test,y_pred))
print('MAPE:',metrics.mean_absolute_percentage_error(y_test,y_pred))
```

```python
feature_name = ['C','cv','cl','fc','pc','da','rd','d','b','w','tn','dt','Av','Al','Vf','lf','df','fv','fl']
```

```python
feature_importances = regressor.feature_importances_
impor_excel = pd.DataFrame([*zip(feature_name,feature_importances)])
impor_excel.sort_values(by=1,ascending=True)
```

```python
#嵌入法，threshold学习曲线
from sklearn.feature_selection import SelectFromModel
score = []
for i in [0.001,0.002,0.003,0.004,0.005,0.006,0.007,0.008,0.009]:
    x_embedded = SelectFromModel(regressor,threshold=i).fit_transform(X_std,y)
    once = cross_val_score(regressor,x_embedded,y,cv=10).mean()
    score.append(once)
plt.plot([0.001,0.002,0.003,0.004,0.005,0.006,0.007,0.008,0.009],score)
plt.show()
#看出0.002为最佳门槛
```

```python
X_embedded = SelectFromModel(regressor,threshold=0.002).fit_transform(X_std,y)
```

```python
#特征选择后的交叉验证跑分
rfc_s = cross_val_score(regressor,X_embedded,y,cv=10).mean()
rfc_s
```
